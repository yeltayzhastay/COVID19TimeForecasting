# COVID-19 бойынша АРИМА моделін болжау

### Библиотекаларды импорттау

```python
from statsmodels.tsa.arima_model import ARIMA
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
import pandas as pd

import os
import warnings
warnings.filterwarnings(action='ignore')
```

### Расталған Жағдайлардың Жалпы Саны

```python
url1 = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
df_confirmed = pd.read_csv(url1)
country = "Kazakhstan"
df_confirmed.head()
```

### Өлім Жағдайларының Жалпы Саны

```python
url2 = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
df_deaths = pd.read_csv(url2)
df_deaths.head()
```

### Емделіп Шыққан Жағдайлардың Жалпы Саны

```python
url3 = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
df_recovered = pd.read_csv(url3)
df_recovered.head()
```

### Визуализациялау

```python
import datetime
import matplotlib.dates as mdates
```

```python
country = 'Kazakhstan'
dates = list(df_confirmed.columns[4:])
dates = list(pd.to_datetime(dates))
dates_kz = dates[8:]

df1 = df_confirmed.groupby('Country/Region').sum().reset_index()
df2 = df_deaths.groupby('Country/Region').sum().reset_index()
df3 = df_recovered.groupby('Country/Region').sum().reset_index()

confirmed = df1[df1['Country/Region'] ==
                country].loc[:, '1/30/20':].values.tolist()[0]
deaths = df2[df2['Country/Region'] ==
             country].loc[:, '1/30/20':].values.tolist()[0]
recovered = df3[df3['Country/Region'] ==
                country].loc[:, '1/30/20':].values.tolist()[0]

active = list(map(int.__sub__, list(
    map(int.__sub__, confirmed, deaths)), recovered))

fig = plt.figure(figsize=(16, 12))
plt.suptitle('Қазақстанда белсенді, сауыққандар, қайтыс болғандар')


plt.plot()
plt.bar(dates_kz, active, color='green', alpha=0.8, label='Белсенді')
plt.bar(dates_kz, recovered, color='grey', label='Сауыққандар')
plt.bar(dates_kz, deaths, color='red', label='Өлім')

plt.title('Қазақстан')
plt.legend()
plt.show()
```

```python
dates = list(df_confirmed.columns[4:])
dates = list(pd.to_datetime(dates))
dates_kz = dates[8:]
```

```python
df1 = df_confirmed.groupby('Country/Region').sum().reset_index()
df2 = df_deaths.groupby('Country/Region').sum().reset_index()
df3 = df_recovered.groupby('Country/Region').sum().reset_index()

k = df1[df1['Country/Region'] == country].loc[:, '1/30/20':]
kz_confirmed = k.values.tolist()[0]

k = df2[df2['Country/Region'] == country].loc[:, '1/30/20':]
kz_deaths = k.values.tolist()[0]

k = df3[df3['Country/Region'] == country].loc[:, '1/30/20':]
kz_recovered = k.values.tolist()[0]

plt.figure(figsize=(15, 10))
plt.xticks(rotation=90, fontsize=11)
plt.yticks(fontsize=10)
plt.xlabel("Уақыты", fontsize=20)
plt.ylabel('Жалпы саны', fontsize=20)
plt.title("Қазақстанда жалпы расталған, белсенді, өлім", fontsize=20)

ax1 = plt.plot_date(y=kz_confirmed, x=dates_kz,
                    label='Расталған', linestyle='-', color='b')
ax2 = plt.plot_date(y=kz_recovered, x=dates_kz,
                    label='Сауыққан', linestyle='-', color='g')
ax3 = plt.plot_date(y=kz_deaths, x=dates_kz, label='Өлім',
                    linestyle='-', color='r')
plt.legend()
```

### Өсу Коэффициентін Есептеу

```python
k = df1[df1['Country/Region'] == country].loc[:, '3/13/20':]
kz_confirmed = k.values.tolist()[0]

growth_diff = []

for i in range(1, len(kz_confirmed)):
    growth_diff.append(kz_confirmed[i] / kz_confirmed[i-1])

growth_factor = sum(growth_diff)/len(growth_diff)
print('Орташа өсу коэффициенті', growth_factor)
```

```python
prediction_dates = []

start_date = dates_kz[len(dates_kz) - 1]
for i in range(15):
    date = start_date + datetime.timedelta(days=1)
    prediction_dates.append(date)
    start_date = date
previous_day_cases = confirmed[len(dates_kz) - 1]
predicted_cases = []

for i in range(15):
    predicted_value = previous_day_cases * growth_factor
    predicted_cases.append(predicted_value)
    previous_day_cases = predicted_value

plt.figure(figsize=(15, 10))
plt.xticks(rotation=90, fontsize=11)
plt.yticks(fontsize=10)
plt.xlabel("Уақыты", fontsize=20)
plt.ylabel('Жалпы саны', fontsize=20)
plt.title("Алдағы 15 күнге болжамды мәндер", fontsize=20)
ax1 = plt.plot_date(y=predicted_cases, x=prediction_dates,
                    linestyle='-', color='c')
```

```python
k = df1[df1['Country/Region'] == country].loc[:, '3/13/20':]
kz_confirmed = k.values.tolist()[0]
data = pd.DataFrame(columns=['ds', 'y'])
data['ds'] = dates[len(dates)-len(kz_confirmed):]
data['y'] = kz_confirmed
```

### Модельді генерациялау және сәйкестендіру

```python
arima = ARIMA(data['y'], order=(5, 1, 0))
arima = arima.fit(trend='c', full_output=True, disp=True)
forecast = arima.forecast(steps=30)
pred = list(forecast[0])

start_date = data['ds'].max()
prediction_dates = []
for i in range(30):
    date = start_date + datetime.timedelta(days=1)
    prediction_dates.append(date)
    start_date = date
plt.figure(figsize=(16, 8))
plt.xlabel("Уақыты", fontsize=20)
plt.ylabel('Жалпы саны', fontsize=20)
plt.title("Келесі 30 күнге болжанатын мәндер", fontsize=20)

plt.plot_date(y=pred, x=prediction_dates, linestyle='dashed',
              color='red', label='Болжанған')
plt.plot_date(y=data['y'], x=data['ds'], linestyle='-',
              color='blue', label='Нақты')
plt.legend()
```

```python
pred_res = []
for i in pred:
    i = int(i)
    pred_res.append(i)
```

```python
arima = ARIMA(data['y'][:-5], order=(5, 1, 0))
arima = arima.fit(trend='c', full_output=True, disp=True)
forecast = arima.forecast(steps=30)
pred1 = list(forecast[0])

start_date1 = data['ds'].iloc[-6]
prediction_dates1 = []
for i in range(30):
    date1 = start_date1 + datetime.timedelta(days=1)
    prediction_dates1.append(date1)
    start_date1 = date1
```

```python
print(arima.summary())
```

### Уақыт Қатарларын Болжау

```python
pred_res = []
for i in pred1:
    i = int(i)
    pred_res.append(i)
```

```python
confirmed_dates = pd.DataFrame()
confirmed_dates['Date'] = data['ds']
confirmed_dates.tail()
```

```python
pred_dates = pd.DataFrame()
pred_dates['Date'] = prediction_dates
pred_dates.head()
```

```python
cc = []
for a in data['y'].iloc[-5:]:
    cc.append(a)
for b in ['NA', 'NA', 'NA', 'NA', 'NA']:
    cc.append(b)
```

```python
dates = []
for i in confirmed_dates['Date'][-5:]:
    dates.append(i)
for j in pred_dates['Date'][:5]:
    dates.append(j)
```

### Модельдің Тиімділігін Бағалау


```python
# Mean Absolute Percentage error(MAPE)
MAPE = np.mean(np.abs(np.array(cc[:5]) -
                      np.array(pred_res[:5]))/np.array(cc[:5]))
print("MAPE-тың мәні = " + str(MAPE*100) + " %")
```

```python
month_pred = pd.DataFrame()
month_pred['Аурудың болжамды жағдайлары 30 күн ішінде'] = pred_res
month_pred
```

```python
r = pd.DataFrame()
r['Уақыты'] = dates
r['Расталған Жағдайлар'] = cc
r['Болжамдалған Жағдайлар'] = pred_res[:10]
r['Модельдің дәлдігі % бойынша'] = str(float(1-MAPE)*100)[:5]
r
```
